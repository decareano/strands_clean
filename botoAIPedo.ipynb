```python
@tool
def get_current_time(timezone: str = "UTC"):
    """Gets the current date and time.
    
    Args:
        timezone: The timezone (currently only UTC supported)
    
    Returns:
        Current date and time as a string
    """
    now = datetime.now()
    #result = f"Current time ({timezone}): {now.strftime('%Y-%m-%d %H:%M:%S')}"
    result = get_current_time("PDT")
    print(result)
    return f"Current time ({timezone}): {now.strftime('%Y-%m-%d %H:%M:%S')}"
    

print("âœ… Time tool created!")
```

```python
import requests
import json

# Test tools that CREATE things (don't need existing IDs)
test_tools = [
    {
        "name": "createCollection",
        "arguments": {
            "collection": {
                "info": {
                    "name": "Test Collection from MCP",
                    "schema": "https://schema.getpostman.com/json/collection/v2.1.0/collection.json"
                },
                "item": []  # Empty collection
            }
        }
    },
    {
        "name": "getWorkspaces",  # Might exist for discovery
        "arguments": {}
    },
    {
        "name": "getCurrentUser",  # Might exist
        "arguments": {}
    }
]

for i, tool in enumerate(test_tools):
    print(f"Testing tool: {tool['name']}")
    response = requests.post(
        "https://mcp.postman.com/mcp",
        headers={
            "Authorization": "blablabla",
            "Content-Type": "application/json", 
            "Accept": "application/json, text/event-stream"
        },
        json={
            "jsonrpc": "2.0",
            "id": i + 1,
            "method": "tools/call",
            "params": tool
        }
    )
    print(f"Status: {response.status_code}")
    if response.status_code == 200:
        print(f"âœ… SUCCESS: {response.text[:300]}...")
    else:
        print(f"Response: {response.text[:200]}...")
    print("---")
```
